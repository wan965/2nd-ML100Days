### Reads TXT files and factor of split imagenet and urls

```python
import pandas as pd

with open('./imagenet_urls_examples.txt', encoding='utf-8') as f:
    image_urls=[line.strip().split('\t') for line in f.readlines()]
```

```python
image_urls
```

### Trans list into Dataframe

```python
DF=pd.DataFrame(image_urls, columns=['imagenet','urls'])
```

```python
DF
```

```python
### show picture throught PIL model, ignore error if a url 
```

```python
first5_urls=DF.iloc[0:5].urls
```

```python
from PIL import Image
import urllib
from io import BytesIO


for url in first5_urls:
    try:
        response = urllib.request.urlopen(url)
        im=Image.open(BytesIO(response.read()))
        #im.show()
        display(im)
    except:
        print("%s is lost" % url)
```
